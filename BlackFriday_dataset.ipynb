```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train=pd.read_csv('train.csv')
df_train.head()
```

```python
df_train.info()
```

```python
df_train.describe()
```

```python
df_test=pd.read_csv('test.csv')
df_test.head()
```

**Merge both the datasets**

```python
df = pd.concat([df_train, df_test], ignore_index=True) # Use concat to combine the DataFrames
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

**Detete the userId column as there is no use of this column for our dataset**

```python
df.drop('User_ID',axis=1,inplace=True)
```

```python
df.columns
```

**Converting the categorical features to numberical**

```python
df['Gender']=df['Gender'].map({'F':0,'M':1})
```

```python
df['Age'].unique()
```

```python
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df.info()
```

```python
df['City_Category'].unique()
```

```python
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
df=pd.concat([df,df_city],axis=1)
df.head()
```

```python
df.drop('City_Category',axis=1,inplace=True)
```

```python
df.info()
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype(int)
```

```python
df.info()
```

**Now all are in numerical datatype so the next step is to handle the missing values**

```python
df.isnull().sum()
```

```python
df['Product_Category_2'].unique()
```

**Both the product category 2 and 3 are discrete feature so we replace the missing values with the mode**

```python
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df['B']=df['B'].astype(int)
df['C']=df['C'].astype(int)
```

```python
df.isnull().sum()
```
