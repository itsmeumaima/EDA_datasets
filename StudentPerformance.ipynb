**First handle the null values**

```python
data.isnull().sum()
```

Replacing the null values present in teacher_quality column with its mode

```python
data['Teacher_Quality'].value_counts()
```

```python
data['Teacher_Quality']=data['Teacher_Quality'].fillna(data['Teacher_Quality'].mode()[0])
```

Do the same thing with the Parental education level for replacing the null values

```python
data['Parental_Education_Level'].value_counts()
```

```python
data['Parental_Education_Level']=data['Parental_Education_Level'].fillna(data['Parental_Education_Level'].mode()[0])
```

Now for the column distance from home

```python
data['Distance_from_Home'].value_counts()
```

```python
data['Distance_from_Home']=data['Distance_from_Home'].fillna(data['Distance_from_Home'].mode()[0])
```

```python
data.isnull().sum()
```

Successfully replace all the null values

```python
data.head()
```

```python
data['Distance_from_Home'].value_counts()
```

Replacing near with 1 , moderate with 2 and far with 3 in the column distance_from_home

```python
data['Distance_from_Home']=data['Distance_from_Home'].map({'Near':1,'Moderate':2,'Far':3})
```

Replacing college with 1 , high school with 2 and postgraduate with 3 in the column Parental_Education_level

```python
data['Parental_Education_Level'].value_counts()
```

```python
data['Parental_Education_Level']=data['Parental_Education_Level'].map({'College':1,'High School':2,'Postgraduate':3})
```

```python
data['Learning_Disabilities']=data['Learning_Disabilities'].map({'Yes':1,'No':0})
```
